```python

import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('research_data.xls') 
print(data)
```

```python
data.info()
```

```python
df = pd.DataFrame(data, columns= ['type', 'solution'])
```

```python
df
```

```python
x=np.array(df['solution']) 
y=np.array(df['type'])
```

```python
x
```

```python
y
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
v = TfidfVectorizer()
x = v.fit_transform(df['solution'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1) 
print(X_train.shape) #- input train features
print(X_test.shape)  #- #input test features
print(y_train.shape) #- #output train labels
print(y_test.shape) #-  #output test labels
```

```python
#importing random forest classifier from assemble module
from sklearn.ensemble import RandomForestClassifier
```

```python
# creating a RF classifier
clf = RandomForestClassifier() 
 
# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
clf.fit(X_train, y_train) 
```

```python
predictions = clf.predict(X_test)
```

```python
#predictions
from sklearn.metrics import mean_squared_error

#%%
# Based on a rule of thumb, it can be said that RMSE values between 0.2 and 0.5 shows that the model can relatively predict the data accurately
mean_squared_error(y_test, predictions)
```

```python
from sklearn.metrics import  accuracy_score
print ('accuracy score= ' , accuracy_score(y_test, predictions))
```

```python
from sklearn.metrics import classification_report
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```
